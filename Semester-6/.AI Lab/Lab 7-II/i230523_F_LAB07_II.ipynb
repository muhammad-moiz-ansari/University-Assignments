```python
import random
import copy

class Item:
  def __init__(self, weight, value):
    self.weight = weight
    self.value = value


# Calculate the fitness value for the chromosome
def calc_fitness(chromo, items):
  fitness = 0
  weight = 0

  for i in range(len(chromo)):
    if chromo[i] == 1:
      fitness += items[i].value
      weight += items[i].weight

  return fitness, weight

def calc_population_fitness(population, items, max_w):
  fitness = []
  pop_copy = copy.deepcopy(population)

  for chromo in pop_copy:
    fit_val, weight = calc_fitness(chromo, items)
    if weight > max_w:
      population.remove(chromo) # discard
    else:
      fitness.append(fit_val)

  return fitness

# The chromosomes with higher fitness have a higher probability of being selected
def selection(pop, fitness):

  # I selected the top best two because they will have higher chance of generating the better chromosomes

  combined = []
  for i in range(len(pop)):
    combined.append((pop[i], fitness[i]))

  # Decending sort so top is best 2
  combined.sort(key=lambda x: x[1], reverse = True)
  chromo1 = combined[0][0]
  chromo2 = combined[1][0]

  return chromo1, chromo2

def crossover(chromo1, chromo2):
  # Perform crossover on the selected chromosomes to create a new offspring
  offspring1 = chromo1[:2] + chromo2[2:]
  offspring2 = chromo2[:2] + chromo1[2:]
  return offspring1, offspring2

def mutation(chromo):
  # Mutate a chromosome
  rand_ind = random.randint(0, len(chromo)-1)
  rand_val = random.randint(0, 1)
  chromo[rand_ind] = rand_val
  pass

# Main function
def main():
  # Items
  items = [Item(15, 30), Item(40, 100), Item(15, 40), Item(20, 50)]

  # Create a random initial population of 20 chromosomes
  max_capacity = 50
  initial_count = 40
  population = [[random.randint(0, 1) for i in range(4)] for j in range(initial_count)]
  print("Initial Population:")
  for pop in population:
    print(pop)

  gen_num = 50

  # MAIN LOOP
  for i in range(gen_num):
    fitness = calc_population_fitness(population, items, max_capacity)

    chromo1, chromo2 = selection(population, fitness)
    offspring1, offspring2 = crossover(chromo1, chromo2)
    population.append(offspring1)
    population.append(offspring2)
    mutation(offspring1)
    mutation(offspring2)
    population.append(offspring1)
    population.append(offspring2)
    population = population[0:initial_count]

  final_fitness = calc_population_fitness(population, items, max_capacity)
  max_fit = max(final_fitness)
  best_idx = final_fitness.index(max_fit)
  best_chromosome = population[best_idx]
  fit_val, max_weight = calc_fitness(best_chromosome, items)

  print('\n--- FINAL RESULTS ---')
  print('Highest Fitness value:', max_fit)
  print('Max Weight: ', max_weight)
  print('Best Chromosome:', best_chromosome)
main()
```
